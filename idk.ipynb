*Step 4 : Data splitting*

```python
from sklearn.model_selection import train_test_split

df_tips_train, df_tips_test = train_test_split(df_tips,test_size=0.2,random_state=28)

df_tips_train_prepro = df_tips_train.copy()
df_tips_test_prepro = df_tips_test.copy()
```

*Step 5 : Prepossessing*

*5.1-Feature Imputation*

*5.1.1-Quantitative Feature Imputation*

*Quantitative Imputation 1 : SimpleImputer*

```python
import numpy as np
from sklearn.impute import SimpleImputer

quantitative_columns= df_tips_train_prepro.select_dtypes(include="number").columns

quantitative_imputer = SimpleImputer(missing_values=np.nan,strategy="median")

df_tips_train_prepro[quantitative_columns] = quantitative_imputer.fit_transform(df_tips_train_prepro[quantitative_columns])
df_tips_test_prepro[quantitative_columns] = quantitative_imputer.transform(df_tips_test_prepro[quantitative_columns])

df_tips_train_prepro.head()
```

*Quantitative Imputation 2 : KNNImputer*

```python
#from sklearn.impute import KNNImputer

#quantitative_columns= df_tips_train_prepro.select_dtypes(include="number").columns

#quantitative_imputer = KNNImputer(n_neighbors=5)

#df_tips_train_prepro[quantitative_columns] = quantitative_imputer.fit_transform(df_tips_train_prepro[quantitative_columns])
#df_tips_test_prepro[quantitative_columns] = quantitative_imputer.transform(df_tips_test_prepro[quantitative_columns])

#df_tips_train_prepro.head()
```

*5.1.2-Qualitative Feature Imputation*

*Qualitative Imputation 1 : SimpleImputer*

```python
import numpy as np
from sklearn.impute import SimpleImputer

qualitative_columns= df_tips_train_prepro.select_dtypes(include=["object", "category", "string"]).columns

qualitative_imputer = SimpleImputer(missing_values=np.nan,strategy="most_frequent")

df_tips_train_prepro[qualitative_columns] = qualitative_imputer.fit_transform(df_tips_train_prepro[qualitative_columns])
df_tips_test_prepro[qualitative_columns] = qualitative_imputer.transform(df_tips_test_prepro[qualitative_columns])

df_tips_train_prepro.head()
```

*5.2-Feature Engineering*

*5.2.1-Feature Creation*

*Variable Creation 1 : Is_weekend_creation*

```python
df_tips_train_prepro["is_weekend_creation"] = df_tips_train_prepro["day"].isin(["Sat", "Sun"]).astype(int)
df_tips_test_prepro["is_weekend_creation"] = df_tips_test_prepro["day"].isin(["Sat", "Sun"]).astype(int)

df_tips_train_prepro.head()
```

*5.2.2-Feature Transformation*

*Variable Transformation 1 : log_total_bill_transformation*

```python
import numpy as np

df_tips_train_prepro["log_total_bill_transformation"] = np.log1p(df_tips_train_prepro["total_bill"])
df_tips_test_prepro["log_total_bill_transformation"] = np.log1p(df_tips_test_prepro["total_bill"])

df_tips_train_prepro.head()
```

*5.2.3-Feature Extraction*

*5.2.3.1- Date Feature Extraction*

*Dans notre dataset, il n’existe pas de variables de type date nécessitant une extraction spécifique.*

*5.2.3.2- String Feature Extraction*

*Dans notre dataset, il n’existe pas de variables de type string nécessitant une extraction spécifique.*

*5.2.4-Feature Interaction*

*Variable Interaction 1 : bill_weekend_interaction*

```python
df_tips_train_prepro["bill_weekend_interaction"] = df_tips_train_prepro["total_bill"] * df_tips_train_prepro["is_weekend_creation"]
df_tips_test_prepro["bill_weekend_interaction"] = df_tips_test_prepro["total_bill"] * df_tips_test_prepro["is_weekend_creation"]

df_tips_train_prepro.head()
```

*5.2.5-Feature Binning*

*Variable Binning 1 : size_binning*

```python
import pandas as pd 

df_tips_train_prepro["size_binning"] = pd.cut(df_tips_train_prepro["size"],bins=[0, 2, 4, 10],labels=["small", "medium", "large"])
df_tips_test_prepro["size_binning"] = pd.cut(df_tips_test_prepro["size"],bins=[0, 2, 4, 10],labels=["small", "medium", "large"])

df_tips_train_prepro.head()
```

*5.3-Feature Encoding*

*5.3.1-Nominal Feature Encoding*

*Nominal Encoding 1 : OneHotEncoder*

```python
from sklearn.preprocessing import OneHotEncoder

qualitative_columns = df_tips_train_prepro.select_dtypes(include=["object","category","string"]).columns
qualitative_columns_nominals = ["sex", "smoker", "day", "time"]

encoder_nominal = OneHotEncoder(sparse_output=False).set_output(transform="pandas")

encoded_train = encoder_nominal.fit_transform(df_tips_train_prepro[qualitative_columns_nominals])
encoded_test = encoder_nominal.transform(df_tips_test_prepro[qualitative_columns_nominals])

df_tips_train_prepro = df_tips_train_prepro.drop(columns=qualitative_columns_nominals).join(encoded_train)
df_tips_test_prepro = df_tips_test_prepro.drop(columns=qualitative_columns_nominals).join(encoded_test)

df_tips_train_prepro.head()
```

*5.3.2-Ordinal Feature Encoding*

*Ordinal Encoding 1 : OrdinalEncoder*

```python
from sklearn.preprocessing import OrdinalEncoder

qualitative_columns = df_tips_train_prepro.select_dtypes(include=["object","category","string"]).columns
qualitative_columns_ordinals = ["size_binning"]

encoder_ordinal = OrdinalEncoder().set_output(transform="pandas")

encoded_train = encoder_ordinal.fit_transform(df_tips_train_prepro[qualitative_columns_ordinals])
encoded_test = encoder_ordinal.transform(df_tips_test_prepro[qualitative_columns_ordinals])

df_tips_train_prepro = df_tips_train_prepro.drop(columns=qualitative_columns_ordinals).join(encoded_train)
df_tips_test_prepro = df_tips_test_prepro.drop(columns=qualitative_columns_ordinals).join(encoded_test)

df_tips_train_prepro.head()
```

*5.4-Feature Scaling*

*Scaling 1 : StandardScaler*

```python
from sklearn.preprocessing import StandardScaler

quantitative_columns=df_tips_train_prepro.select_dtypes(include="number").columns

scaler_standard = StandardScaler()

df_tips_train_prepro[quantitative_columns] = scaler_standard.fit_transform(df_tips_train_prepro[quantitative_columns])
df_tips_test_prepro[quantitative_columns] = scaler_standard.transform(df_tips_test_prepro[quantitative_columns])

df_tips_train_prepro.head()
```

*Step 6 : Modeling*


*6.1- Model Preparation*

```python
X_train = df_tips_train_prepro.drop("tip", axis=1)
y_train = df_tips_train_prepro["tip"]

X_test = df_tips_test_prepro.drop("tip", axis=1)
y_test = df_tips_test_prepro["tip"]
```

*6.2 – Model Training*

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

*4.2 – Model Evaluation*

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE :", mae)
print("RMSE :", rmse)
print("R2 :", r2)
```
